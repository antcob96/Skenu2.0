# Satellite Orbirt

```python
# Importing numpy and matplotlib
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
filename = 'Orbit.dat'

all_data = np.loadtxt(filename, skiprows=0)
all_data = all_data.transpose()

x = all_data[0]
y = all_data[1]
```

```python

f = plt.figure()
plt.plot(x,y, 'black', linewidth = 0.4, label = 'Satellite')


plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.xlabel('x')
plt.ylabel('y')


plt.show()
f.savefig("SatelliteOrbit.pdf", format = 'pdf', bbox_inches = "tight")
```

```python
filename = 'Ratio_semi_maj_min.dat'

all_datar = np.loadtxt(filename, skiprows=0)
all_datar = all_datar.transpose()

k = all_datar[0]
ratio = all_datar[1]
```

```python

f = plt.figure()
plt.plot(k,ratio, 'black', linewidth = 1.0)



plt.xlabel('Number of orbits, n')
plt.ylabel('Ratio of semimajor and semiminor, b/a')


plt.show()
f.savefig("Ratio.pdf", format = 'pdf', bbox_inches = "tight")
```
